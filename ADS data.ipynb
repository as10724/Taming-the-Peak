```python
import pylab as pl
import pandas as pd
import numpy as np
%pylab inline

import os
import csv
```

```python
M_I_In = pd.read_csv('ACS_14_5YR_B01003_with_ann.csv')
M_I_Po = pd.read_csv('ACS_14_5YR_B19013_with_ann.csv')
all_data = pd.merge(M_I_In, M_I_Po, on='GEO.id2')
all_data.head()
```

```python
all_data.describe()
```

```python
all_data = all_data[['GEO.id2', 'HD01_VD01_y', 'HD01_VD01_x']]
all_data.head()
```

```python
all_data.columns = [['ZipCode', 'MedianIncome', 'Population']]
all_data.columns
```

```python
M_I_Zp = pd.read_excel('zipcodes.xlsx')
M_I_Zp.describe()
```

```python
all_data['ZipCode'] = all_data['ZipCode'].convert_objects(convert_numeric=True)
M_I_Zp['ZipCode'] = M_I_Zp['ZipCode'].convert_objects(convert_numeric=True)
```

```python
data = all_data[all_data.ZipCode.isin(M_I_Zp.ZipCode)]
```
