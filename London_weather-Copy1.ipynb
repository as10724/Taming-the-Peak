```python
weather_clean_11
```

```python
weather_clean_11_r = pd.DataFrame()
for i in range(len(weather_clean_11)):
    if weather_clean_11['Datetime'][i][14:] == '20:00<br />':
        weather_clean_11_r = weather_clean_11_r.append(weather_clean_11.iloc[(i)])
weather_clean_11_r.reset_index(inplace=True)
```

```python
weather_clean_11_r
```

```python
weather_clean_11_r.drop('index', 1, inplace=True)
for i in range(len(weather_clean_11_r)):
    weather_clean_11_r['Datetime'][i] = weather_clean_11_r['Datetime'][i][:14] + '00:00'    
weather_clean_11_r
```

```python
weather_clean_11_r.to_csv('London_weather_hourly_winter_2011-12_(00).csv')
```
