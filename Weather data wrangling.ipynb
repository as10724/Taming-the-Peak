```python
from __future__ import print_function
import pandas as pd, os, numpy as np
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim()
def getzip(a):
   location = geolocator.reverse(a)
   return (location.raw['address']['postcode'])
```

```python
weather = pd.read_csv("Weather Data.csv")
```

```python
weather = weather[weather['LATITUDE'] != 'unknown']
weather = weather[weather['LONGITUDE'] != 'unknown']
weather.head()
```

```python
coord = weather[['LATITUDE', 'LONGITUDE']]
coord = pd.DataFrame.drop_duplicates(coord)
zips = []
for i in range(len(coord)):
    try:
         zips.append(getzip(str(str(coord['LATITUDE'][i]) + "," + str(coord['LONGITUDE'][i]))))
    except KeyError: 
        zips.append('NA')
coord['ZipCode'] = zips
```
